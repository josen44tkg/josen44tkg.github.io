```
import numpy as np
import matplotlib.pyplot as plt
from random import seed
from random import  random
import math as mt
```

Determinamos el numero de datos de cada Clase , en este caso la clase minoritaria es t2 la cual tendrá la mitad de datos que t1.

De igual manera determinamos el radio de la Clase 1

```
#seed()
print("numero de datos clase 1")
n = int(0 +(random()*(1000-0)))#int(input())
t1 = np.zeros((n,3))
print(n)
print("numero de datos calse 2")
n = int(n/2) 
#int(100 +(random()*(10000-100)))#int(input())
t2 = np.zeros((n,3))
print(n)
print("tamaño radio mayor")
n = 100#int(input())

```

Se generan datos aleatoreos de la clase 2 utilizando random


```



for i in range(0,t1.shape[0]):
  a =random() * 2 * mt.pi
  r = n * mt.sqrt(random())
  t1[i][2]=i
  while r<=(n/2)+10:
    r = n * mt.sqrt(random())
  t1[i][0]=int(r * mt.cos(a))
  t1[i][1]=int(r * mt.sin(a))


for i in range(0,t2.shape[0]):
  a =random() * 2 * mt.pi
  r = n/2 * mt.sqrt(random())
  t2[i][0]=int(r * mt.cos(a))
  t2[i][1]=int(r * mt.sin(a))
  t2[i][2]=i
```

```
plt.scatter(np.transpose(t1)[0],np.transpose(t1)[1],marker ='x',color=(0.2,0.1,0.4))
plt.scatter(np.transpose(t2)[0],np.transpose(t2)[1],marker ='o',color=(0.8,0.1,0.4))
```

```
cent_t1 = np.zeros((2,1))
cent_t2 = np.zeros((2,1))

for i in range(0,t1.shape[0]):
  cent_t1[0] = cent_t1[0] + t1[i][0]
  cent_t1[1] = cent_t1[1] + t1[i][1]



for i in range(0,t2.shape[0]):
  cent_t2[0] = cent_t2[0] + t2[i][0]
  cent_t2[1] = cent_t2[1] + t2[i][1]


cent_t1 =cent_t1/t1.shape[0]


cent_t2 =cent_t2/t2.shape[0]
```
